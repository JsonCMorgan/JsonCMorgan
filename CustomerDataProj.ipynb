# Categorizing Customers into Age Groups

To facilitate a more detailed analysis of our customers based on their age, we categorize them into specific age groups. This categorization helps in understanding the distribution of customers across different age ranges and can provide valuable insights for targeted marketing strategies.

## Steps in the Process:

1. **Defining Age Bins and Labels**:
   - I defined a series of age bins, which are ranges of ages. These bins will be used to group customers.
   - Corresponding to each age bin, I defined a set of labels that represent these age groups.

```python
# Define age bins and labels
age_bins = [0, 18, 25, 35, 45, 55, 65, 100]
age_labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66-100']


````python
# Define age bins and labels
age_bins = [0, 18, 25, 35, 45, 55, 65, 100]
age_labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66-100']

# Create a new column for age groups
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Display the first few rows to verify
df.head()
````
