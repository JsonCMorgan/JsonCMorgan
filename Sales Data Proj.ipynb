# Enhancing the Dataset with Time-Based Insights

- An integral part of my data analysis strategy involves extracting meaningful information from existing data. In this step, I aimed to enrich our sales dataset by adding time-based insights, which are pivotal in understanding buying patterns and customer behavior during different days of the week.

```python
# Ensure the invoice_date is in datetime format
sales_data['invoice_date'] = pd.to_datetime(sales_data['invoice_date'])

# Create 'is_weekend' column: 5 and 6 correspond to Saturday and Sunday
sales_data['is_weekend'] = sales_data['invoice_date'].dt.dayofweek >= 5

```

# Rationale Behind This Enhancement:

## Datetime Conversion: 

- Ensuring invoice_date is in the correct datetime format is crucial for any time-based analysis. This format allows for the extraction of valuable information like the day of the week, month, and year from the date.

## Weekend Analysis: 

- By creating the is_weekend column, I could easily differentiate between weekend and weekday transactions. This binary categorization (where Saturday and Sunday are considered weekends) sets the stage for further analysis, such as comparing weekend sales with weekday sales.

This data enrichment step is instrumental in preparing the dataset for deeper analyses. By understanding when our customers are more likely to make purchases, we can tailor our marketing efforts, adjust staffing levels, and manage inventory more effectively.

# Statistical Comparison of Weekend vs. Weekday Sales

- In my quest to derive deeper insights from our sales data, I utilized a statistical approach to compare sales performance between weekends and weekdays. A t-test, a fundamental tool in statistical analysis, was employed to determine whether there's a significant difference in sales between these two timeframes.

```python
from scipy.stats import ttest_ind

# Separate the sales data into weekend and weekday
weekend_sales = sales_data[sales_data['is_weekend']]['price']
weekday_sales = sales_data[~sales_data['is_weekend']]['price']

# Perform t-test
t_stat, p_val = ttest_ind(weekend_sales, weekday_sales)
print("t-statistic:", t_stat)
print("p-value:", p_val)
```

# Understanding the T-Test Results:

## T-Statistic: 

- The t-statistic reflects the size of the difference relative to the variation in our sample data. A higher t-statistic value indicates a more significant difference between the groups.

## P-Value: 

- This value tells us about the probability of obtaining the observed results, assuming that there is no significant difference between the weekend and weekday sales.

- A low p-value (typically < 0.05) suggests that the observed difference is statistically significant.

- A high p-value implies that any difference in sales might be due to random chance.

# Strategic Implications:

## Data-Driven Decision Making: 

- These results empower us to make informed decisions based on statistical evidence. For instance, if the difference in sales is significant, we might consider different strategies for weekends and weekdays.

## Customer Behavior Insights: 

- Understanding when customers are more likely to make purchases can influence marketing strategies, operational planning, and resource allocation.

This analysis exemplifies how statistical methods can be applied to real-world business data to extract meaningful insights. It highlights my approach to using data-driven evidence to inform business strategies.

# Visualizing Seasonal Sales Trends by Financial Quarter

- A pivotal part of my analysis involved understanding how sales fluctuated throughout the year. By breaking down sales data by month and correlating it with financial quarters, I aimed to identify seasonal patterns and trends that could inform strategic planning and marketing efforts.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import calendar

# Ensure invoice_date is a datetime
sales_data['invoice_date'] = pd.to_datetime(sales_data['invoice_date'])

# Extract month
sales_data['month'] = sales_data['invoice_date'].dt.month

# Aggregate sales by month
monthly_sales = sales_data.groupby('month')['price'].sum()

# Convert month numbers to abbreviations
monthly_sales.index = [calendar.month_abbr[i] for i in monthly_sales.index]

# Define colors for each financial quarter
colors = ['blue' if month in ['Jan', 'Feb', 'Mar'] else
          'green' if month in ['Apr', 'May', 'Jun'] else
          'orange' if month in ['Jul', 'Aug', 'Sep'] else
          'red' for month in monthly_sales.index]

# Plotting
plt.figure(figsize=(12, 6))
ax = monthly_sales.plot(kind='bar', color=colors)
plt.title('Seasonal Sales Analysis by Financial Quarter', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

# Set y-axis limit from 4,000,000 to 9,000,000
ax.set_ylim([4000000, 9000000])

# Format y-axis labels to represent millions
y_labels = ['{:.1f}M'.format(y/1000000) for y in ax.get_yticks()]
ax.set_yticklabels(y_labels)

plt.xticks(rotation=0)
plt.show()
```

# Segmenting Customers by Spending Behavior

- To deepen my understanding of customer spending habits, I segmented our customers based on their total sales. This segmentation provides insights into the distribution of customer spending and helps identify our most valuable customers.

```python
# Aggregate total sales per customer
customer_total_sales = sales_data.groupby('customer_id')['price'].sum()

# Sort customers by total sales
sorted_customers = customer_total_sales.sort_values(ascending=False)

# You can segment further, for example, into high, medium, and low spending customers
high_spending_customers = sorted_customers[sorted_customers > sorted_customers.quantile(0.75)]
medium_spending_customers = sorted_customers[(sorted_customers <= sorted_customers.quantile(0.75)) & 
                                             (sorted_customers > sorted_customers.quantile(0.25))]
low_spending_customers = sorted_customers[sorted_customers <= sorted_customers.quantile(0.25)]

# Output the count of customers in each segment
print(f"High Spending Customers: {len(high_spending_customers)}")
print(f"Medium Spending Customers: {len(medium_spending_customers)}")
print(f"Low Spending Customers: {len(low_spending_customers)}")
```

# Cohort Analysis

- In a Cohort Analysis, customers are grouped based on their first interaction or purchase with the business. You then track these groups over time to analyze behaviors like retention, repeat purchase rate, or average spending.

## Steps to Define Cohorts:

- Identify the First Purchase Date for Each Customer:

- This involves finding the earliest purchase date for each customer, which will be used to define the cohorts.

## Assign Each Customer to a Cohort:

- Customers are grouped into cohorts based on the month or period of their first purchase.

## Prepare the Data for Analysis:

- Organize the data so that each customer's subsequent behavior can be tracked in relation to their cohort.

```python
import pandas as pd

# Ensure invoice_date is in datetime format
sales_data['invoice_date'] = pd.to_datetime(sales_data['invoice_date'])

# Find the first purchase date for each customer
first_purchase = sales_data.groupby('customer_id')['invoice_date'].min().reset_index()
first_purchase.columns = ['customer_id', 'first_purchase_date']

# Extract the month from the first purchase date
first_purchase['cohort'] = first_purchase['first_purchase_date'].dt.to_period('M')

# Merge the cohort data with the original sales data
sales_data = pd.merge(sales_data, first_purchase[['customer_id', 'cohort']], on='customer_id')

# Check the first few entries of the sales data to ensure the merge was successful
print(sales_data.head())
```

# Integrating Cohort Analysis into Sales Data

- In an effort to gain deeper insights into our customer acquisition and retention patterns, I embarked on a cohort analysis. This technique groups customers based on their first purchase date, allowing us to observe how specific customer cohorts behave over time.

# Monthly Retention Analysis:

## Calculate Retention Rate:

- Determine the number of unique customers in each cohort who made purchases in subsequent months.

## Prepare a Retention Matrix:

- Create a matrix where each row represents a cohort, and each column represents the months since the first purchase.

```python
# Create a column for the invoice month
sales_data['invoice_month'] = sales_data['invoice_date'].dt.to_period('M')

# Group by cohort and invoice month and count unique customers
grouped = sales_data.groupby(['cohort', 'invoice_month']).agg(n_customers=('customer_id', 'nunique')).reset_index()

# Create a period number column (months since first purchase)
grouped['period_number'] = (grouped.invoice_month - grouped.cohort).apply(lambda x: x.n)

# Pivot the data to create the retention matrix
retention_matrix = grouped.pivot_table(index='cohort', columns='period_number', values='n_customers')

# Calculate retention rate
cohort_size = retention_matrix.iloc[:,0]
retention_matrix = retention_matrix.divide(cohort_size, axis=0)

# Display the retention matrix
print(retention_matrix.head())
```

# Insights and Benefits of Cohort Analysis:

## Cohort Identification: 

- By identifying the first purchase date for each customer and categorizing them into cohorts based on this date, we can track and compare the behavior of different customer groups over time.

## Data Enrichment: 

- Adding a 'cohort' column to our sales data enriches the dataset, allowing for more sophisticated analyses, such as customer lifetime value and retention rates.

## Effective Marketing Strategies: 

- Understanding how different cohorts behave helps in tailoring marketing campaigns and customer engagement strategies to various segments.

This step of integrating cohort analysis into our sales data is a significant move towards a more nuanced understanding of our customer base. It lays the groundwork for advanced analyses that can inform strategic decisions and drive business growth.

# Verifying Repeat Purchase Flags in Sales Data

- In my continuous effort to dissect customer purchasing patterns, I implemented a method to identify repeat purchases within our sales data. This process is crucial in distinguishing between one-time and returning customers, offering insights into customer loyalty and repeat purchasing behaviors.

```python
# Verify the repeat purchase flag
sales_data['is_repeat'] = sales_data['invoice_date'] > sales_data['first_purchase_date']
print("Repeat Purchases Flag:\n", sales_data['is_repeat'].value_counts())
```


# Significance of the Repeat Purchase Flag:

## Flagging Repeat Purchases: 

- By comparing each transaction's invoice_date with the customer's first_purchase_date, I was able to flag transactions that occurred after the first purchase. This flag serves as an indicator of repeat purchases.
